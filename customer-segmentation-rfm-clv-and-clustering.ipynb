Customer Personality Analysis is a detailed analysis of a company‚Äôs ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviours and concerns of different types of customers.

Customer personality analysis helps a business modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company‚Äôs database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

# **Soluation Architecture**
"Analytics Maturity Model" or the "Analytics Continuum". It represents the progressive stages of data analysis in increasing order of complexity and value:


- **Description Analytics:** Descriptive analytics is the foundational level of data analysis. It involves summarizing historical data to provide insights into what has happened in the past. This often includes basic statistics, visualizations, and summary reports. Descriptive analytics answers questions like "What happened?" and "What is the current state of affairs?"
Diagnostic Analytics:

- **Description Analytics:** Diagnostic analytics goes a step beyond descriptive analytics. It aims to understand why certain events or trends occurred in the past. This involves exploring relationships between variables, identifying patterns, and uncovering the root causes of specific outcomes. Diagnostic analytics answers questions like "Why did it happen?" and "What were the contributing factors?"
Predictive Analytics:

- **Description Analytics:** Predictive analytics uses historical data and statistical algorithms to make informed predictions about future events or trends. It involves building models that can forecast or estimate outcomes based on patterns observed in the data. Predictive analytics answers questions like "What is likely to happen in the future?" and "What are the potential outcomes?"
Prescriptive Analytics:

- **Prescriptive Analytics:** Prescriptive analytics is the highest level of data analysis. It not only predicts future outcomes but also provides recommendations or actions that should be taken to achieve a desired outcome. This involves using optimization and simulation techniques to determine the best course of action. Prescriptive analytics answers questions like "What should we do to achieve a specific goal?"

```python
# Convert 'Dt_Customer' to datetime format
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], format='%d-%m-%Y')
# Define the date '29/6/2015' and convert it to datetime
specific_date = pd.to_datetime('29/6/2015', format='%d/%m/%Y')
# Calculate the difference between each date in the column and the specific date
data['Time_Since_Enrollmen'] = (specific_date - data['Dt_Customer']).dt.days

# print minimun and maximum Date on the dataset
display("Data timeline {} - {}".format(data["Dt_Customer"].min(), data["Dt_Customer"].max()))
```

- Generate the **'NumCmpResponse'** column by aggregating the counts from 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', and 'Response'. This column represents the total number of responses across various marketing campaigns.

- Define the **'Churn'** column based on a business rule, considering any customer who has not made any purchases in the last 60 days as 'churned'

## **Features Engineering**<a id="4.1"></a>

```python
# Calculate the total amount spent on products by summing up the individual product categories.
# This creates a new column named 'TotalMntProducts'.
data["TotalPurchaseAmount"] = data[["MntWines", "MntFruits", "MntMeatProducts",
                                 "MntFishProducts", "MntSweetProducts",
                                 "MntGoldProds"]].sum(axis=1)


# Calculate the total number of purchases made by summing up the different purchase channels.
# This creates a new column named 'TotalNumPurchases'.
data["TotalPurchases"] = data[["NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases"]].sum(axis=1)



# Calculate the total number of campaign responses by summing up the individual campaign acceptance columns.
# This creates a new column named 'NumCmpResponse'.
data["NumCmpResponce"] = data[["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3",
                                 "AcceptedCmp4", "AcceptedCmp5", "Response"]].sum(axis=1)


# 0 for not churn, 1 for low churn, and 2 for churn
conditions = [
    data['Recency'] <= 30, # not churn
    (data['Recency'] > 30) & (data['Recency'] < 60), # low churn
    data['Recency'] >= 60 # churned
]
categories = [0, 1, 2]

# Apply the conditions and categories using numpy.select()
data['Churn'] = np.select(conditions, categories, default=-1)
```

## **CLV: Customer Lifetime Value** <a id="4.2"></a>

Customer lifetime value (CLV) is a measure of the total income a business can expect to bring in from a typical customer for as long as that person or account remains a client.

<img src="https://popupsmart.com/encyclopedia/user/pages/26.customer-lifetime-value-clv/customer-lifetime-value.jpg">

```python
# Calculate Average Purchase Value (APV)
data['APV'] = round(data['TotalPurchaseAmount'] / data['TotalPurchases'],3)
# Replace infinite values with 0
data.loc[np.isinf(data["APV"]), "APV"] = 0


# Calculate Average Purchase Frequency (APF)
data['APF'] = round(data['TotalPurchases'] / data['Time_Since_Enrollmen'],3)

# Calculate customer lifespan
# Average Number Deals Purchase  + Avergae Number Responce of campaign
lifespan = round(data["NumDealsPurchases"].mean() + data["NumCmpResponce"].mean())

# Calculate CLV
data['CLV'] = round(data['APV'] * data['APF'] * lifespan,3)

data["CLV_Score"] = pd.qcut(data['CLV'], q=3, labels=['Low', 'Medium', 'High'])
```

## **RFM (Recency, Frequency and  Monetary)** <a id="4.3"></a>

RFM is a method used for segmenting customers based on their past behavior.
It looks at three key dimensions:
 - Recency (how recently a customer made a purchase)
 - Frequency (how often a customer makes a purchase)
 - Monetary Value (how much money a customer spends)

<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*RkpRpnvXJAocXOSv9H1txw.jpeg">

```python
# Define scoring criteria for each RFM value
recency_scores = [5, 4, 3, 2, 1]  # Higher score for lower recency (more recent)
frequency_scores = [1, 2, 3, 4, 5]  # Higher score for higher frequency
monetary_scores = [1, 2, 3, 4, 5]  # High
bins=5
# Calculate RFM scores
data['RecencyScore'] = pd.cut(data['Recency'], bins=bins, labels=recency_scores)
data['FrequencyScore'] = pd.cut(data['TotalPurchases'], bins=bins, labels=frequency_scores)
data['MonetaryScore'] = pd.cut(data['TotalPurchaseAmount'], bins=bins, labels=monetary_scores)

# Convert RFM scores to numeric type
data['RecencyScore'] = data['RecencyScore'].astype(int)
data['FrequencyScore'] = data['FrequencyScore'].astype(int)
data['MonetaryScore'] = data['MonetaryScore'].astype(int)

# Calculate RFM score by combining the individual scores
data['RFM_Score'] = data['RecencyScore'] + data['FrequencyScore'] + data['MonetaryScore']

# Create RFM segments based on the RFM score
segment_labels = ['Low-Value', 'Mid-Value', 'High-Value']
data['Value_Segment'] = pd.qcut(data['RFM_Score'], q=3, labels=segment_labels)
data[["ID" ,'Value_Segment']].sample(3)

# Create a new column for RFM Customer Segments
data['RFM_Customer_Segments'] = ''

# Assign RFM segments based on the RFM score
data.loc[data['RFM_Score'] >= 9, 'RFM_Customer_Segments'] = 'Best Customers'
data.loc[(data['RFM_Score'] >= 6) & (data['RFM_Score'] < 9), 'RFM_Customer_Segments'] = 'Loyal Customers'
data.loc[(data['RFM_Score'] >= 5) & (data['RFM_Score'] < 6), 'RFM_Customer_Segments'] = "At Risk Customers"
data.loc[(data['RFM_Score'] >= 4) & (data['RFM_Score'] < 5), 'RFM_Customer_Segments'] = "Needs Attention"
data.loc[(data['RFM_Score'] >= 3) & (data['RFM_Score'] < 4), 'RFM_Customer_Segments'] = "Lost Customers"

# Print the updated data with RFM segments
display(data[['ID', 'RFM_Customer_Segments']].sample(3))
```

### ***Labels Encoding*** <a id="5.2.1"></a>

```python
def Labels_Encoder(data_series:pd.Series) -> tuple:

    # Initialize the label encoder
    label_encoder = LabelEncoder()

    # Apply label encoding to the data Series
    data_series = label_encoder.fit_transform(data_series)

    # Return a tuple containing the label encoded Series and the LabelEncoder object
    return data_series, label_encoder
```

```python
# Apply label encoding to the 'Education' and Marital_Status columns using the custom Labels_Encoder function.
# This line updates the columns with label encoded values and also stores the LabelEncoder objects.
X["Education"], Education_Labels_Encoder = Labels_Encoder(X["Education"])
#X["Marital_Status"], Marital_Status_Labels_Encoder = Labels_Encoder(X["Marital_Status"])
X["Partnered_Status"], RFM_Score_Labels_Encoder = Labels_Encoder(X["Partnered_Status"])
X["RFM_Score"], RFM_Score_Labels_Encoder = Labels_Encoder(X["RFM_Score"])
display(X.head(3))
```

```python
cluster_data = pd.concat([X_standardized[["RFM_Score", "CLV"]], pca_results], axis=1)
cluster_data.sample(3)
```

```python
# Create an instance of KElbowVisualizer with KMeans model (n_init="auto") and range of clusters (k=10)
plt_elb = KElbowVisualizer(KMeans(init = 'k-means++',random_state = 42,n_init=10), k=10)
# Fit the KMeans model to the cluster_data and perform the elbow method to find the optimal number of clusters
plt_elb.fit(cluster_data)
# Display the elbow plot which helps to determine the optimal number of clusters
plt_elb.show();
```

```python
# We run K-means with a fixed number of clusters. In our case 3.
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
# We divide our data into the four clusters.
kmeans.fit(cluster_data)

# We create a new data frame with the original features and add a new column with the assigned clusters for each point.
data.loc[:,'Cluster'] = kmeans.labels_
merged_df.loc[:,'Cluster'] = kmeans.labels_

# Define a mapping from cluster numbers to labels
cluster_mapping = {0: 'cluster one', 1: 'cluster two', 2: 'cluster three', 3: 'cluster four'}
# Replace the values in the 'cluster' column with their corresponding labels
data['Cluster_Str'] = data['Cluster'].replace(cluster_mapping)
merged_df['Cluster_Str'] = merged_df['Cluster'].replace(cluster_mapping)
```
